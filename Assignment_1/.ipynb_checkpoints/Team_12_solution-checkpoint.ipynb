<h4>Problem 2</h4>
<p>With the price and capacity data provided for each dock, Charles would like you to
find out what is the optimal distribution across the different options for next year.
He is expecting you to provide the exact numbers, as well as the expected costs that
would result from your proposal.</p>

<b>Constrained cost minimisation problem is as follows:</b>

$
\begin{aligned}
    \min_{Q,P} \quad & Q \cdot P = q_1 \cdot p_1 \ + \ ... \\
    \textrm{s.t.} \quad & \sum_{q\in{Q}}{q}\cdot 66 \cdot 0.45 = 1500000 \\
      &Q \le  Max\_Capacity \\
      &Q \ge 0 \\
\end{aligned}
$

```python
P, M = df["40_ft_container_price_eur"].values, df["max_capacity"].values
```

```python
# create a list of target variables
Q = [
    pulp.LpVariable(dock, lowBound=0, cat=pulp.LpContinuous) for dock in df.dock
]
```

```python
objective_func = pulp.lpSum(
    [q*p for q, p in zip(Q, P)]
)
```

```python
problem = pulp.LpProblem("cost_minimisation", pulp.LpMinimize)

problem += objective_func

for quantity, max_capacity in zip(Q, M):
    problem += quantity <= max_capacity
    
problem += pulp.lpSum([q*66*0.45 for q in Q]) == 1_500_000
```

```python
# solve optimisation problem
problem.solve(
    pulp.PULP_CBC_CMD(msg=False)
)
```

```python
# Solution definition
solution_2 = {
    "variables": {
        var.name: var.value() for var in problem.variables()
    },
    "objective_value": problem.objective.value()
}

solution_2
```

<p>Convert real solution into whole number using ceiling function:</p>

```python
def integerize_solution(solution):
    integer_solution = {}

    integer_solution["variables"] = {
        key: math.ceil(val) for key, val in solution["variables"].items()
    }

    integer_solution["objective_value"] = sum([
        value * df.loc[df.dock == key, "40_ft_container_price_eur"].values[0]
        for key, value in integer_solution["variables"].items()
    ])

    return integer_solution
```

```python
integerize_solution(solution_2)
```

<p>We can just ceil the quantity with decimal points because solution itself is found as follows: solver is trying to fill in demand of 1.5 mil tons using cheapest options, so once the cheapest option is used up in terms of max capacity it goes to the next cheapest one. Therefore, we can just assume that ceiling the result is more or less the same as output of integer solver apart from the fact that we artifically expanded the import constraint a little bit to allow any solution for the problem.</p>

<h4>Problem 3</h4>
<p>The following docks have a sign-up fee. This means that, if Beanie Limited wants to
do any business with them at all, they need to pay this fee. This fee is only paid once
and is independent of the container volume that goes through the dock.</p>

<ul>
    <li>Algeciras: 800,000€</li>
    <li>Marseille: 500,000€</li>
    <li>Antwerp: 1,000,000€</li>
</ul>

<p>To avoid logistics netfork being so complex maximum number of docks is now no more than <b>3</b></p>

<b>Constrained cost minimisation problem is as follows:</b>

$
\begin{aligned}
    \min_{Q,P} \quad & Q \cdot P + A \cdot 1 = q_1 \cdot p_1 \ + sign\_up_1 \cdot is\_used_1\ ... \\
    \textrm{s.t.} \quad & \sum_{q\in{Q}}{q}\cdot 66 \cdot 0.45 = 1500000 \\
      &Q \le  Max\_Capacity \cdot A \\
      &\sum_{a\in{A}}{a} \le 3 \\ 
      &Q \ge 0 \\
\end{aligned}
$

```python
# define dictionary of additional fees for starting a contract with dock
ADDITIONAL_COSTS = {
    "Rotterdam": 0,
    "Antwerp": 1_000_000,
    "Hamburg": 0,
    "Amsterdam": 0,
    "Marseille": 500_000,
    "Algeciras": 800_000,
    "Valencia": 0,
    "Genoa": 0
}

A = ADDITIONAL_COSTS.values()
```

```python
def solve_problem(
    Q: List[pulp.LpVariable], 
    P: List[int], 
    M: List[int],
    at_most: int = 3
) -> Dict[str, Any]:
    
    problem = pulp.LpProblem("cost_minimisation", pulp.LpMinimize)
    
    # dot product of prices and quantities
    objective_func = pulp.lpSum([q*p for q, p in zip(Q, P)])
    # add sign-up costs
    objective_func += pulp.lpSum([i*m for i, m in zip(I, A)])
    
    problem += objective_func
    
    # add constraints
    for quantity, max_capacity, is_used in zip(Q, M, I):
        problem += quantity <= is_used * max_capacity
    
    problem += pulp.lpSum([q*66*0.45 for q in Q]) == 1_500_000
    problem += pulp.lpSum(I) <= at_most
    
    problem.solve(pulp.PULP_CBC_CMD(msg=False))
    
    return {
        "variables": {
            var.name: var.value() for var in problem.variables() if "is_used" not in var.name
        },
        "objective_value": problem.objective.value()
    }
```

```python
Q = [
    pulp.LpVariable(dock, lowBound=0, cat=pulp.LpContinuous) for dock in df.dock
]

I = [
    pulp.LpVariable(f"{dock}_is_used", lowBound=0, cat=pulp.LpBinary) for dock in df.dock
]
```

```python
# Real number solution
solution_3 = solve_problem(Q, P, M, at_most=3)
solution_3
```

```python
sol = integerize_solution(solution_3)
sol
```

```python
18_177_820 - 16_392_820
```

<p>As a result, we observe that costs are slightly higher than in previous case if we are to use only 3 docks. If the amount we save by using 5 docks as in previous case instead of 3 is greater than additional costs coming from additional management and transportation costs of more complex logistics network, then we should definitely go for solution with 5 docks, otherwise this one is optimal</p>

<h4>Problem 4</h4>

$Valencia\_price \sim{U[310, 390]}$ </br>
$Algeciras\_price \sim{U[280, 330]}$

<p>As we were discussing during the break, there is not enough variation in prices to yield any other unique solutions. If you change right boundaries to any bigger number new solutions will pop up, that could be analysed and discussed.</p>

```python
from functools import partial
```

```python
RANDOM_PRICES = {
    "Valencia": partial(np.random.uniform, 310, 1000),
    "Algeciras": partial(np.random.uniform, 280, 1000)
}
```

```python
def get_prices() -> List[float]:
    prices = []
    for i, dock in enumerate(df.dock):
        if dock in RANDOM_PRICES:
            prices.append(RANDOM_PRICES[dock]())
        else:
            prices.append(df["40_ft_container_price_eur"][i])
    return prices
```

```python
solutions = []

for _ in tqdm(range(10000)):
    
    P = get_prices()
    solution = solve_problem(Q=Q, P=P, M=M, at_most=5)
    solution.update({
        "sampled_prices": P
    })
    
    solutions.append(solution)
```

```python
df_sim = pd.DataFrame.from_dict([
    {**solution["variables"], "objective_value": solution["objective_value"]} for solution in solutions
])
```

```python
mean_values = {
    col: [] for col in df_sim.columns
}

for col in df_sim.columns:
    for i in range(1, df_sim.shape[0]+1):
         mean_values[col].append(
             df_sim[col][:i].mean()
         )
```

```python
df_mean = pd.DataFrame(mean_values)

df_mean.plot(
    subplots=True, figsize=(16, 9), layout=(5, 2)
)

plt.show()
```

<h4>Stability of solution matrix</h4>

```python
# for each preset of prices assign unique optimal solution
unique_solutions = []

for solution in solutions:
    if solution["variables"] not in unique_solutions:
        unique_solutions.append(solution["variables"])
```

```python
solution_map, unique_solution_map = [], {}

for solution in solutions:
    solution_idx = unique_solutions.index(solution["variables"])
    if solution_idx not in unique_solution_map:
        unique_solution_map[solution_idx] = solution["variables"]
    solution_map.append({
        "objective_value": solution["objective_value"],
        "solution_idx": solution_idx,
        "sampled_prices": solution["sampled_prices"]
    })
```

```python
sampled_prices = np.array([
    (solution["sampled_prices"][5], solution["sampled_prices"][6]) for solution in solution_map
])
solution_indecies = [
    solution["solution_idx"] for solution in solution_map
]
objective_values = [
    solution["objective_value"] for solution in solution_map
]

fig, (axs) = plt.subplots(1, 2, figsize=(16, 6))
ax1, ax2 = axs

ax1.scatter(sampled_prices[:, 0], sampled_prices[:, 1], c=solution_indecies)
ax1.set_xlabel("Sampled Valencia Price")
ax1.set_ylabel("Sampled Algeciras Price")
ax1.set_title("Optimal Solution type")

ax2.scatter(
    sampled_prices[:, 0], sampled_prices[:, 1], c=objective_values, cmap=plt.cm.Blues
)
ax2.set_xlabel("Sampled Valencia Price")
ax2.set_ylabel("Sampled Algeciras Price")
ax2.set_title("Optimal Total Costs")

plt.show()
```

```python
def calculate_costs(solution: Dict[str, float], prices: List[float]) -> float:
    X = np.array(list(solution.values()))
    P = np.array(prices)
    total_cost = np.dot(X, P)

    for dock in solution:
        if dock in ADDITIONAL_COSTS:
            total_cost += ADDITIONAL_COSTS[dock]
    
    return total_cost
```

```python
solution_costs_sampled = {
    f"solution_{i}": [] for i in range(5)
}

for _ in tqdm(range(100000)):
    P = get_prices()
    for i in range(5):
        solution_costs_sampled[f"solution_{i}"].append(
            calculate_costs(unique_solution_map[i], P)
        )
```

```python
df_sim_costs = pd.DataFrame(solution_costs_sampled)

df_sim_costs.plot.hist(
    bins=50, subplots=True, figsize=(16, 10), layout=(3, 2)
)

plt.show()
```

```python
def q5(x):
    return x.quantile(.05)
def q95(x):
    return x.quantile(.95)

df_sim_costs.agg([q5, "median", q95, "mean"]).style.background_gradient(axis=1)
```
